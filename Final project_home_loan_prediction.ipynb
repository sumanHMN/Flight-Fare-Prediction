### Data Cleaning and Preprocessing

#### Feature Engineering

```python
# Feature Engineering: Creating new features
application_train['INCOME_CREDIT_RATIO'] = application_train['AMT_INCOME_TOTAL'] / application_train['AMT_CREDIT']
application_train['CREDIT_ANNUITY_RATIO'] = application_train['AMT_CREDIT'] / application_train['AMT_ANNUITY']
application_train['INCOME_PER_PERSON'] = application_train['AMT_INCOME_TOTAL'] / (application_train['CNT_CHILDREN'] + 1)

# Check the new features
print(application_train[['INCOME_CREDIT_RATIO', 'CREDIT_ANNUITY_RATIO', 'INCOME_PER_PERSON']].head())
```

### One-Hot Encoding

```python
# One-Hot Encoding for categorical features
categorical_columns = application_train.select_dtypes(include=['object']).columns
application_train = pd.get_dummies(application_train, columns=categorical_columns, drop_first=True)

# Check the transformed dataset
print(application_train.head())
```

#### Scaling and Normalization

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Define numerical features to scale
numeric_features = application_train.select_dtypes(include=['number']).columns

# Initialize scalers
minmax_scaler = MinMaxScaler()
standard_scaler = StandardScaler()

# Apply Min-Max Scaling
application_train_minmax_scaled = application_train.copy()
application_train_minmax_scaled[numeric_features] = minmax_scaler.fit_transform(application_train[numeric_features])

# Apply Standard Scaling
application_train_standard_scaled = application_train.copy()
application_train_standard_scaled[numeric_features] = standard_scaler.fit_transform(application_train[numeric_features])

# Check the scaled datasets
print(application_train_minmax_scaled.head())
print(application_train_standard_scaled.head())
```

### Data Integration

#### Aggregating Supplementary Data

```python
# Aggregating bureau data
bureau_agg = bureau.groupby('SK_ID_CURR').agg({
    'AMT_CREDIT_SUM': ['sum', 'mean', 'max'],
    'AMT_CREDIT_SUM_DEBT': ['sum', 'mean', 'max'],
    'DAYS_CREDIT': ['mean', 'max']
}).reset_index()

# Flatten column names
bureau_agg.columns = ['SK_ID_CURR'] + ['BUREAU_' + '_'.join(col).upper() for col in bureau_agg.columns[1:]]

# Merge with application_train
application_train = application_train.merge(bureau_agg, on='SK_ID_CURR', how='left')
```

#### POS_CASH_balance Data Aggregation

```python
# Aggregating POS_CASH_balance data
pos_cash_agg = POS_CASH_balance.groupby('SK_ID_CURR').agg({
    'MONTHS_BALANCE': ['max', 'mean'],
    'SK_DPD': ['sum', 'max'],
    'SK_DPD_DEF': ['sum', 'max']
}).reset_index()

# Flatten column names
pos_cash_agg.columns = ['SK_ID_CURR'] + ['POS_' + '_'.join(col).upper() for col in pos_cash_agg.columns[1:]]

# Merge with application_train
application_train = application_train.merge(pos_cash_agg, on='SK_ID_CURR', how='left')
```

```python
# Aggregating credit_card_balance data
credit_card_agg = credit_card_balance.groupby('SK_ID_CURR').agg({
    'AMT_BALANCE': ['sum', 'mean', 'max'],
    'AMT_CREDIT_LIMIT_ACTUAL': ['mean'],
    'SK_DPD': ['sum', 'max']
}).reset_index()

# Flatten column names
credit_card_agg.columns = ['SK_ID_CURR'] + ['CREDIT_' + '_'.join(col).upper() for col in credit_card_agg.columns[1:]]

# Merge with application_train
application_train = application_train.merge(credit_card_agg, on='SK_ID_CURR', how='left')
```

```python
# Aggregating previous_application data
previous_app_agg = previous_application.groupby('SK_ID_CURR').agg({
    'AMT_CREDIT': ['sum', 'mean', 'max'],
    'AMT_ANNUITY': ['mean'],
    'DAYS_FIRST_DUE': ['min'],
    'DAYS_TERMINATION': ['max']
}).reset_index()

# Flatten column names
previous_app_agg.columns = ['SK_ID_CURR'] + ['PREV_' + '_'.join(col).upper() for col in previous_app_agg.columns[1:]]

# Merge with application_train
application_train = application_train.merge(previous_app_agg, on='SK_ID_CURR', how='left')
```

```python
# Fill missing values with 0 for aggregated fields
application_train.fillna(0, inplace=True)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, classification_report
```

```python
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, classification_report

# Initialize results dictionary
results = {}

# Evaluate models
for model_name, model in models.items():
    try:
        # Train the model
        model.fit(X_train_scaled, y_train)

        # Predictions
        y_pred = model.predict(X_test_scaled)

        # Probability Predictions
        y_pred_proba = None
        roc_auc = None
        if hasattr(model, "predict_proba"):
            y_pred_proba = model.predict_proba(X_test_scaled)[:, 1]
            roc_auc = roc_auc_score(y_test, y_pred_proba)

        # Metrics
        accuracy = accuracy_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred)

        # Store results
        results[model_name] = {
            "Accuracy": accuracy,
            "F1 Score": f1,
            "ROC AUC": roc_auc,
        }

        # Print results
        roc_auc_str = f"{roc_auc:.4f}" if roc_auc is not None else "N/A"
        print(f"\n{model_name} Results:")
        print(f"Accuracy: {accuracy:.4f}, F1 Score: {f1:.4f}, ROC AUC: {roc_auc_str}")
        print(classification_report(y_test, y_pred))

    except Exception as e:
        print(f"Error with model {model_name}: {e}")

# Display results in a DataFrame
results_df = pd.DataFrame(results).T
print("\nModel Performance:\n", results_df)
```

### Model Evaluation

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, classification_report

# Evaluate models
evaluation_results = {}

for model_name, model in models.items():
    # Predictions
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else None

    # Metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred_proba) if y_pred_proba is not None else None

    # Store results
    evaluation_results[model_name] = {
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1 Score": f1,
        "ROC AUC": roc_auc
    }

    # Print classification report for detailed insights
    print(f"\n{model_name} Classification Report:")
    print(classification_report(y_test, y_pred))
```

```python
# Create a DataFrame for easy comparison
evaluation_df = pd.DataFrame(evaluation_results).T

# Display the evaluation report
print("\nModel Performance Comparison:")
print(evaluation_df)
```
